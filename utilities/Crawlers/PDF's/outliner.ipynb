### 讀出標題
- 先將所有內容按照跳行指令將其切割成一行行文字的序列
- 每行的前導文字是否含有標題編號，如是，則挑出成為outlines
- 沒有妥善命名的標題、只是用編號來寫一段段的文字內容，此處還是將其視為標題，以60字為限。希望前面幾個字能出現重點(要靠語言模型了)。

```python
a=''
i=1
for pageObj in pdfReader.pages:
    a+=pageObj.extract_text().replace('7-'+str(i)+' ','')
    i+=1
lines=a.split('\n')
outlines=[l.split(':')[0].split('：')[0].strip() for l in lines]
outlines=[l for l in outlines if (dots(l) or CNnum(l) or ENnum(l))  and len(l)<60]
outlines=[l for l in outlines if '-' not in l]
for i in outlines[:30]:
    print(i)
```

- 這支副程式在確認前導的文字是否含有1.、(1)、（1）等等情況。

```python
def dots(l):
    tdic={'':'.','(':')','（':'）'}
    for p in ['','(','(']:
        for i in range(1,11):
            a=p+str(i)+tdic[p]
            n=len(a)
            if l[:n]==a:return True
    return False    
```

- 這2組副程式的邏輯是一樣的，除了中間的英數字之外，前後括弧也是比對的重點。

```python
def CNnum(l):
    num=['一','二','三','四','五','六','七','八','九','十',]
    tdic={'':'、','(':')','（':'）'}
    for p in ['','(','(']:
        for i in num:
            a=p+i+tdic[p]
            n=len(a)
            if l[:n]==a:return True
    return False    
```

```python
def ENnum(l):
    num='ABCDEFGHabcdefgh'
    tdic={'':'.','(':')','（':'）'}
    for p in ['','(']:
        for i in num:
            t=''
            if p in [' ','']:t='.'
            a=p+i+tdic[p]
            n=len(a)
            if l[:n]==a:return True
    return False   
```

### 開啟標題檔
- 開啟文字檔
- 儲存成pandas的工作表

```python
from pandas import *
fname="D:\E\My Documents\EIAs\電廠\GAS_EIA\(1070401A) 興達電廠燃氣機組更新改建計畫環境影響說明書\C07outlineR.txt"
with open(fname,'r') as f:
    lines=[l.strip().replace('\n','') for l in f]
df=DataFrame({'tit':lines,'lev':0})    
```

- 按照標題前的數字(文字)編號，辨識其階層。以便讓後續層級間關係認定會更好操作。
- 因第1與第2層太相像，先解決第2層，在給定第1層，這樣的邏輯較為容易。

```python
chnum=['一','二','三','四','五','六','七','八','九','十',]
df['id']=df.index
df.loc[df.tit.map(lambda x:x.count('.')==2 and x[0].isdigit() and x[2].isdigit() and x[4].isdigit()),'lev']=2
df.loc[(df.lev==0)&(df.tit.map(lambda x:x[2].isdigit() and x[1]=='.')),'lev']=1
df.loc[(df.lev==0)&(df.tit.map(lambda x:x[0] in chnum)),'lev']=3
df.loc[(df.lev==0)&(df.tit.map(lambda x:x[1] in chnum)),'lev']=4
df.loc[(df.lev==0)&(df.tit.map(lambda x:x[0].isdigit())),'lev']=5
df.loc[(df.lev==0)&(df.tit.map(lambda x:x[1].isdigit() and x[0]=='(' and x[2]==')')),'lev']=6
df.loc[(df.lev==0)&(df.tit.map(lambda x:x[0].isupper())),'lev']=7
df.loc[(df.lev==0)&(df.tit.map(lambda x:x[1].isupper() and x[0]=='(' and x[2]==')')),'lev']=8
df.loc[(df.lev==0)&(df.tit.map(lambda x:x[0].islower())),'lev']=9
df.loc[(df.lev==0)&(df.tit.map(lambda x:x[1].islower() and x[0]=='(' and x[2]==')')),'lev']=10
```

最後的程式寫成這樣。雖然比較笨、但還蠻管用的。解釋一下:
- `mnlev,mxlev`是階層的最高和最低層(數字小的是高層)。這樣寫可以適用在所有情況，也不限定一定要從0開始。
- 每一(母)層先從df中篩出來(`a`)，因為outline是按照順序的，所以`a`注標之間就是該段的範圍。做出起訖注標(`ai,aj`)邏輯就更清楚。
- 子層(`b`)的定義。考量到很多同仁並不按照階層規範逐一編寫，會有跳層的情形，所以此處篩出凡是階層數低於`a`、且尚未給定從屬關係(`df.parent_id`)的內容。
- 按照前述起訖注標(`ai,aj`)分段篩選子層(`b`)，挑出當中的最高層，將其與母層進行連結，就是將`df.parent_id`令為`ai[i]`。

```python
df['parent_id'] = None
mnlev,mxlev=df.lev.min(),df.lev.max()
for l in range(mnlev,mxlev):
    a=df.loc[df.lev==l]
    n=len(a)
    if n==0:continue
    ai=list(a.index)
    aj=ai[1:]+[len(df)]
    b=df.loc[(df.lev>l) & (df.parent_id.isnull())] 
    for i in range(n):
        boo=(b.index>ai[i]) & (b.index<aj[i] )
        c=b.loc[boo]
        mn=c.lev.min()
        df.loc[c.loc[c.lev==mn].index,'parent_id']=ai[i]        
```

```python
import PyPDF2
from pandas import *
fname='C07.pdf'
pdfFileObj = open(fname, 'rb')
pdfReader = PyPDF2.PdfReader(pdfFileObj)
a=''
i=1
if '7-' not in pdfReader.pages[0].extract_text():i=0
if '7 -' not in pdfReader.pages[0].extract_text():i=0
for pageObj in pdfReader.pages:
    a+=pageObj.extract_text().replace('7-'+str(i),'\n').replace('7 - '+str(i),'\n')
    i+=1
lines=a.split('\n')
nlines=len(lines)
```

```python
df=read_csv('outlineR.csv')
n=len(df)
df['content']=''
iold=0
for i in df.index[:-1]:
    for j in range(iold,nlines):
        if df.tit[i+1] in lines[j]:
            for jj in range(iold,j):
                df.content[i]+=lines[jj].strip()
            df.content[i].replace(df.tit[i],'')
            iold=j
            break
for jj in range(iold,nlines):
    df.content[n-1]+=lines[jj].strip()
```
