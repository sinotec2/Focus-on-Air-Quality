```python
def FloatNotInLine(line):
    list=line.split()
    for j in list:
        try:
            i=float(j)
            return False
        except:
            continue
    return True
```

```python
def CharNotAhead(l):
    if '月' in l[:3]:return False
    if '%' in l[:4]:return False
    if '％' in l[:4]:return False
    return True
```

```python
def dots(l):
    tdic={'':'.','(':')','（':'）'}
    for p in ['','(','(']:
        for i in range(1,11):
            a=p+str(i)+tdic[p]
            n=len(a)
            if l[:n]==a:return True
    return False    
```

```python
def CNnum(l):
    num=['一','二','三','四','五','六','七','八','九','十',]
    tdic={'':'、','(':')','（':'）'}
    for p in ['','(','(']:
        for i in num:
            a=p+i+tdic[p]
            n=len(a)
            if l[:n]==a:return True
    return False    
```

```python
def ENnum(l):
    num='ABCDEFGHabcdefgh'
    tdic={'':'.','(':')','（':'）'}
    for p in ['','(']:
        for i in num:
            t=''
            if p in [' ','']:t='.'
            a=p+i+tdic[p]
            n=len(a)
            if l[:n]==a:return True
    return False   
```
